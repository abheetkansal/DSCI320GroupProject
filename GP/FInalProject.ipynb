```python
import pandas as pd
import altair as alt

dataSetA = pd.read_csv('world.csv')
dataSetC = pd.read_csv('earthquake_1995-2023.csv')

# selected_columns = df.loc[:, ['Column1', 'Column2', 'Column3']]

dataSetCfiltered = dataSetC.loc[:, ['magnitude', 'date_time', 'country']]
dataSetCfiltered_by_country = dataSetCfiltered.groupby(['country'])['magnitude'].max()
dataSetA.columns = dataSetA.columns.str.upper()

# Merge the DataFrames based on the 'Country' column
merged_data2 = pd.merge(dataSetCfiltered, dataSetCfiltered_by_country, on='country', how = 'left')

merged_data3 = merged_data2.drop('magnitude_x', axis=1)
merged_data3.columns = merged_data3.columns.str.upper()

merge_data4 = pd.merge(dataSetA, merged_data3, on='COUNTRY',how='left')
m5_filtered = merge_data4.drop(['ABBREVIATION', 'LARGEST CITY','CALLING CODE','GROSS TERTIARY EDUCATION ENROLLMENT (%)'], axis = 1)
#This code it to see all the Columns we have available to us.
pd.set_option('display.max_columns', None)
m5_filtered.head()
data = m5_filtered

data.head()
```

```python
# We are now adding ordinal data that focuses on The Global Perceptions of Countries such as Corruption and Generosity
whr = pd.read_csv("WHR2023.csv")

#Converting the data to different Levels that we can use.

labels = ['Very Low', 'Low', 'Moderate', 'High', 'Very High']
levels = [1,2,3,4,5]
whr['Level Of Corruption'] = pd.cut(whr['Perceptions of corruption'], bins=5, labels = labels)
whr['Level Of Social Support in Society'] = pd.cut(whr['Social support'], bins=5, labels = labels)
whr['Generosity'] = pd.cut(whr['Generosity'], bins=5, labels = levels)
whr['Freedom to make life choices'] = pd.cut(whr['Freedom to make life choices'], bins=5, labels = levels)
whr['Ladder score'] = pd.cut(whr['Ladder score'], bins=5, labels = levels)

#Renaming Columns so we can have a common scheme

whr['COUNTRY'] = whr['Country name']
whr_filtered = whr[['COUNTRY', 'Generosity','Ladder score', 'Freedom to make life choices', 'Level Of Social Support in Society', 'Level Of Corruption']] 

#merging together all of our filtered data

data = pd.merge(data, whr_filtered, on='COUNTRY', how = 'left')
data.head()
```

```python
# Reformating all of our chosen columns so we can do EDA

data['date'] = pd.to_datetime(data['DATE_TIME']).dt.date
data['density'] = data['DENSITY\n(P/KM2)']
data['GDP'] = data['GDP'].str.replace(',', '').str.replace(' ', '').str.replace('$', '').astype(float)
data['CO2-EMISSIONS'] = data['CO2-EMISSIONS'].str.replace(',', '').str.replace(' ', '')
data['FORESTED AREA (%)'] = data['FORESTED AREA (%)'].str.replace('%', '').str.replace(' ', '')
data['POPULATION'] = data['POPULATION'].str.replace(',', '').str.replace(' ', '')
data['TOTAL TAX RATE'] = data['TOTAL TAX RATE'].str.replace('%', '').str.replace(' ', '')
data['UNEMPLOYMENT RATE'] = data['UNEMPLOYMENT RATE'].str.replace('%', '').str.replace(' ', '')
data['MINIMUM WAGE'] = data['MINIMUM WAGE'].str.replace('$', '').astype(float)
data['density'] = pd.to_numeric(data['density'], errors='coerce')
data['FORESTED AREA (%)'] = pd.to_numeric(data['FORESTED AREA (%)'], errors='coerce')
data['PHYSICIANS PER THOUSAND'] = pd.to_numeric(data['PHYSICIANS PER THOUSAND'], errors='coerce')
data['TOTAL TAX RATE'] = pd.to_numeric(data['TOTAL TAX RATE'], errors='coerce')
data['UNEMPLOYMENT RATE'] = pd.to_numeric(data['UNEMPLOYMENT RATE'], errors='coerce')
data['CO2-EMISSIONS'] = pd.to_numeric(data['CO2-EMISSIONS'], errors='coerce')
data['POPULATION'] = pd.to_numeric(data['POPULATION'], errors='coerce')
```

```python
data = data[['LONGITUDE', 'LATITUDE','date', 'MAGNITUDE_Y', 'COUNTRY', 'CAPITAL/MAJOR CITY','CURRENCY-CODE', 'density', 'BIRTH RATE','CO2-EMISSIONS', 'FERTILITY RATE', 'FORESTED AREA (%)', 'GDP', 'INFANT MORTALITY',  'LIFE EXPECTANCY', 'MINIMUM WAGE', 'OFFICIAL LANGUAGE','PHYSICIANS PER THOUSAND', 'POPULATION', 'TOTAL TAX RATE', 'UNEMPLOYMENT RATE', 'Generosity', 'Ladder score','Freedom to make life choices', 'Level Of Social Support in Society', 'Level Of Corruption']]
data.head()
```

```python
data = data.drop_duplicates(subset= 'COUNTRY', keep= 'first')
data.to_csv('final_dataset', index=True)
final_dataset = pd.read_csv("final_dataset")
final_dataset = final_dataset.drop(final_dataset.columns[0], axis = 1)
final_dataset
```

```python
density_hist = alt.Chart(final_dataset).mark_bar().encode(
    alt.X('density:Q', bin=alt.BinParams(maxbins=20)),
    alt.Y('count()')
).properties(
    title = 'Population Density of the World'
)
birth_rate_hist = alt.Chart(final_dataset).mark_bar().encode(
    alt.X('BIRTH RATE:Q', bin=alt.BinParams(maxbins=12)),
    alt.Y('count()')
).properties(
    title = 'The Birth Rate of the World'
)
co2_hist = alt.Chart(final_dataset).mark_bar().encode(
    alt.X('CO2-EMISSIONS:Q', bin=alt.BinParams(maxbins=12)),
    alt.Y('count()')
).properties(
    title = 'What do CO2 Emissions look like Overall?'
)
fort_hist = alt.Chart(final_dataset).mark_bar().encode(
    alt.X('FORESTED AREA (%):Q', bin=alt.BinParams(maxbins=12)),
    alt.Y('count()')
).properties(
    title = 'What does the Forested Rate Look Like Overall ?'
)
gdp_plot = alt.Chart(final_dataset).mark_boxplot().encode(
    alt.X('Level Of Corruption'),
    alt.Y('GDP_PER_CAPITA')
).properties(width=400, height=300,
             title = 'How Does the Corruption Level Realate to the GDP per Capita')

corruptionvsunemployment_plot = alt.Chart(final_dataset).mark_boxplot().encode(
    alt.X('Level Of Corruption'),
    alt.Y('UNEMPLOYMENT RATE')
).properties(width=400, height=300,
             title = 'How does Level of Corruption Realte to Unemployment Rate' )

country_plot = alt.Chart(final_dataset).mark_circle().encode(
    alt.X('COUNTRY'),
    alt.Y('UNEMPLOYMENT RATE'),
    alt.Color('OFFICIAL LANGUAGE')
).properties(
    title = 'How the Unemplyment Rate of Countries differs and its effect on Official Languages'
)



country_gdp_plot = alt.Chart(final_dataset).mark_bar().encode(
    alt.X('COUNTRY'),
    alt.Y('GDP_PER_CAPITA', title = 'GDP per Capita'),
    alt.Color('UNEMPLOYMENT RATE')
).properties(
    title = 'GDP per capita of different nations and their unemployment rates'
)

```

```python
alt.vconcat(
    alt.hconcat(density_hist.properties(width=400, height=300), 
                birth_rate_hist.properties(width=400, height=300)),
    alt.hconcat(
        co2_hist.properties(width=400, height=300),
        fort_hist.properties(width=400, height=300)  # Set width and height as needed
    ),
    title='Histogram Dashboard'
).resolve_legend(
    color='independent'
).configure_axis(
    labelAngle=0
)
```

```python
alt.vconcat(
    alt.hconcat(country_gdp_plot.encode(alt.X('COUNTRY', axis = alt.Axis(labelAngle=-45)))),
    alt.hconcat(
        gdp_plot.properties(width=400, height=300)  # Set width and height as needed
    ),
    alt.hconcat(country_plot.encode(alt.X('COUNTRY', axis = alt.Axis(labelAngle=-45))),),
    title='Univariate Plot Dashboard'
).resolve_legend(
    color='independent'
).configure_axis(
    labelAngle=0
)
```

```python
alt.vconcat(
    alt.hconcat(density_hist.properties(width=400, height=300), 
                birth_rate_hist.properties(width=400, height=300)),
    alt.hconcat(
        corruptionvsunemployment_plot.properties(width=400, height=300),
        gdp_plot.properties(width=400, height=300)  # Set width and height as needed
    ),
    title='Some Plots'
).resolve_legend(
    color='independent'
).configure_axis(
    labelAngle=0
)
```

```python
generosity_gdp = alt.Chart(final_dataset).mark_bar().encode(
    alt.X('Generosity:O'),
    alt.Y('Level Of Corruption:O'),
    alt.Color('GDP_PER_CAPITA')
).properties(
    title = 'Looking at Generosity and Levels of Corruption and how they relate to GDP per Capita'
)

gdp_tax = alt.Chart(final_dataset).mark_circle().encode(
    alt.X('GDP_PER_CAPITA:Q').bin(),
    alt.Color('COUNTRY'),
    alt
    .Y('TOTAL TAX RATE:Q').bin()
).properties(
    title = 'Looking at GDP per Capitas of Different Countries and their Total Tax Rates'
)

gdp_tax_box = alt.Chart(final_dataset).mark_boxplot().encode(
    x=alt.X('TOTAL TAX RATE', title='Tax Rate', scale=alt.Scale(domain=[20, 85])),
    y=alt.Y('GDP_PER_CAPITA:Q', title='GDP per Capita'),
    color=alt.Color('TOTAL TAX RATE:N', legend=alt.Legend(title='Tax Rate')),
    tooltip=['TOTAL TAX RATE:N', 'GDPpercapita:Q']
).properties(
    title = 'The Relationship between Total Tax Rate and GDP Per Capita'
)

forest_co2 = alt.Chart(final_dataset).mark_bar().encode(
    alt.X('FORESTED AREA (%)'),
    alt.Y('CO2-EMISSIONS'),
    alt.Color('density')
).properties(
    title = 'The Relationship between CO2 Emissions and Forested area, by density'
)
```

```python

import geopandas as gpd

# load data
gdf_quakies = gpd.read_file(data.earthquakes.url, driver="GeoJSON")
gdf_world = gpd.read_file(data.world_110m.url, driver="TopoJSON")

# defintion for interactive brush
brush = alt.selection_interval(
    encodings=["longitude"],
    empty=False,
    value={"longitude": [-50, -110]}
)

# world disk
sphere = alt.Chart(alt.sphere()).mark_geoshape(
    fill="transparent", stroke="lightgray", strokeWidth=1
)

# countries as shapes
world = alt.Chart(gdf_world).mark_geoshape(
    fill="lightgray", stroke="white", strokeWidth=0.1
)

# earthquakes as dots on map
quakes = alt.Chart(gdf_quakies).transform_calculate(
    lon="datum.geometry.coordinates[0]",
    lat="datum.geometry.coordinates[1]",
).mark_circle(opacity=0.35, tooltip=True).encode(
    longitude="lon:Q",
    latitude="lat:Q",
    color=alt.condition(brush, alt.value("goldenrod"), alt.value("steelblue")),
    size=alt.Size("mag:Q").scale(type="pow", range=[1, 1000], domain=[0, 7], exponent=4),
).add_params(brush)

# combine layers for the map
left_map = alt.layer(sphere, world, quakes).project(type="mercator")

# histogram of binned earthquakes
bars = alt.Chart(gdf_quakies).mark_bar().encode(
    x=alt.X("mag:Q").bin(extent=[0,7]),
    y="count(mag):Q",
    color=alt.value("steelblue")
)

# filtered earthquakes
bars_overlay = bars.encode(color=alt.value("goldenrod")).transform_filter(brush)

# combine layers for histogram
right_bars = alt.layer(bars, bars_overlay)

# vertical concatenate map and bars
left_map | right_bars
```
